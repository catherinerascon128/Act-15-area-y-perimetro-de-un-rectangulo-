1. Estructura funcional:
El código sigue un enfoque procedimental, evitando la programación orientada a objetos para mantener la lógica lineal y fácil de rastrear. Se definen tres funciones principales:
arearectangulo() y perimetrorectangulo(): cada una solicita independientemente la base y la altura al momento de ejecutarse, aplica la fórmula correspondiente y muestra el resultado formateado a dos decimales.
menu(): actúa como el cerebro del programa. Presenta las opciones disponibles y, mediante un ciclo while, mantiene la aplicación activa hasta que el usuario decida salir.
2. Control de flujo:
Dentro de menu() se utiliza la estructura condicional if - elif - else para gestionar la navegación:
Las opciones 1 y 2 invocan directamente a sus funciones de cálculo.
La opción 3 imprime un mensaje de cierre y ejecuta break para finalizar el bucle while.
El bloque else maneja cualquier carácter o número distinto a las opciones válidas, mostrando una advertencia sin detener el programa (cumpliendo con la indicación de no implementar validación de entrada).
3. Interactividad y diseño:
Se emplean emojis y marcos de texto ASCII para organizar visualmente la interfaz, lo que resulta útil para principiantes que asocian mejor la estructura del programa. Además, cada resultado imprime la fórmula aplicada, reforzando el vínculo entre el código y la teoría matemática.
4. Cómo ejecutarlo:
Guarda el código en un archivo llamado geometria_rectangulo.py.
Abre tu terminal o consola de comandos.
Ejecuta: python geometria_rectangulo.py (o python3 según tu sistema operativo).
Interactúa con el menú. No se requieren librerías externas; todo funciona con la instalación base de Python.

```python
def arearectangulo():
    """Calcula el área de un rectángulo a partir de base y altura."""
    print("\n🟦 CÁLCULO DEL ÁREA DEL RECTÁNGULO")
    base = float(input("Ingresa la base: 📏 "))
    altura = float(input("Ingresa la altura: 📏 "))
    area = base * altura
    print(f"✅ Área = base × altura = {area:.2f} unidades²\n")

def perimetrorectangulo():
    """Calcula el perímetro de un rectángulo a partir de base y altura."""
    print("\n🟧 CÁLCULO DEL PERÍMETRO DEL RECTÁNGULO")
    base = float(input("Ingresa la base: 📏 "))
    altura = float(input("Ingresa la altura: 📏 "))
    perimetro = 2 * (base + altura)
    print(f"✅ Perímetro = 2 × (base + altura) = {perimetro:.2f} unidades\n")

def menu():
    """Muestra el menú interactivo y controla el flujo del programa."""
    while True:
        print("╔════════════════════════════════════════╗")
        print("║       📐 MENÚ RECTÁNGULO 📐           ║")
        print("╠════════════════════════════════════════╣")
        print("║ 1️⃣ Calcular área del rectángulo      ║")
        print("║ 2️⃣ Calcular perímetro del rectángulo ║")
        print("║ 3️⃣ Salir del programa                ║")
        print("╚════════════════════════════════════════╝")

        opcion = input("👉 Selecciona una opción (1-3): ")

        if opcion == "1":
            arearectangulo()
        elif opcion == "2":
            perimetrorectangulo()
        elif opcion == "3":
            print("\n👋 ¡Gracias por usar el programa! Sigue practicando. 📚✨")
            break
        else:
            print("\n⚠️ Opción no reconocida. Por favor, elige 1, 2 o 3.\n")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
